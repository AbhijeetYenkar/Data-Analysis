3. Which product categories show the highest frequency of returns?

```python
total_returns = df.groupby('Product_Category')['Return_Flag'].sum()
print(total_returns)
```

```python
plt.bar(total_returns.index,total_returns.values,color= 'orange')
plt.title('Highest Frequency of Returns Across all Categories')
plt.xlabel('Product Category')
plt.ylabel('Return Flag')
plt.show()
```

- There are Returns complaints by order categories.
- Beauty category have 300 return complaints over all orders.
- Clothing category have highest returns complaints more than 550.
- Electronics category have nearly 500 complaints over all orders.
- Grocery category are lowest returns complaints nearly 200, over all orders.
- Home category have more than 350 return complaints over all orders.
