## Q6. Given a DataFrame with dates and temperatures:

Group by Month

For each group, calculate the range (max - min) of temperature using a custom function. 

```python
import pandas as pd
import numpy as np
df = pd.DataFrame({
    'Date': pd.date_range('2022-01-01', periods=100),
    'Temp': np.random.randint(15, 35, size=100)
})
df['Month'] = df['Date'].dt.month
```

```python
gro = df.groupby('Month')['Temp'].min()
print(gro,'Min Temperature Of Months')
```

```python
gro = df.groupby('Month')['Temp'].max()
print(gro,'Max Temperature Of Months')
```

## Q10. For the following dataframe of transactions Plot the total amount spent by each customer using a horizontal bar chart (ordered by value).

```python
df = pd.DataFrame({
    'CustomerID': [101, 102, 101, 103, 102, 104],
    'Amount': [150, 90, 200, 300, 120, 80],
    'TransactionDate': pd.to_datetime([
        '2023-01-01', '2023-01-02', '2023-01-10', 
        '2023-02-01', '2023-02-10', '2023-03-01'
    ]),
    'Region' : ['West','North','East','North','West','West'],
    'Sales' : [525,658,458,795,236,111]
})
          
```

```python
region = df.groupby('Region')['Sales'].sum()
print(region)
```

```python
c = df.groupby('CustomerID')['Amount'].sum()
print(c)
```

```python
import matplotlib.pyplot as plt
plt.barh(c.index,c.values,color='lime')
plt.title('Total Amount Spent By Each Customer')
plt.xlabel('Amount')
plt.ylabel('Customer Id')
plt.show()
```

## Q11. For the dataframe of transactions (above mentioned df).
- Calculate total sales by region

- Plot as a pie chart with percentages and exploded slices

```python
region = df.groupby('Region')['Amount'].sum()
print(region)
```

```python
plt.pie(region, labels=region.index, autopct='%1.1f%%', startangle=90)
plt.title('Total Sales By Region')
plt.axis('equal') 
plt.show()
```
