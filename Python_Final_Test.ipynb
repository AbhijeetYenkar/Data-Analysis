Q.9

```python
lst = [1,2,3,4,5,6,7,8,9]
def prime(x):
    if x < 2:
        return False
    for i in range(2,int(x ** 0.5) + 1):
        if x % i == 0:
            return False
    return x
result = list(filter(lambda a : prime(a) , lst))
print(result)
```

```python
from functools import reduce
lst = [1,2,3,4,5,6,7,8,92,5,5,4,5]
square = list(map(lambda x : x ** 2,lst))
#print(square)
total = reduce(lambda x,y : x + y,square)
print(f'Sum of squares of all elements in list = {total}')
```
